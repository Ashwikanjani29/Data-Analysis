```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Horticulture Area Production Yield Value for Aromatic Plant Crops 2020-21.csv")
df.info()
```

 provides a summary of the dataset, including the number of non-null values, data types of each column, and memory usage.

```python
df.head()
```

displays the first five rows of the dataset, giving a quick preview of its structure and values.

```python
df = df.drop(columns=["Sl No"])
df.head()
```

 removes the "Sl No" column, and confirms the change by displaying the first five rows without it.

```python
df.info()
```

shows the dataset's structure, including column names, data types, non-null counts, and memory usage.

```python
df.columns = [ "District_Name", "Area_Hectares", "Production_MT", "Yield_MT_per_Hectare", "Value_Lakhs"]
df.describe()
```

provides statistical summaries of numerical columns, including count, mean, min, max, and quartiles.

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(df["Area_Hectares"], bins=10, kde=True, color="blue")
plt.title("Distribution of Area (Hectares)")
plt.xlabel("Area (Hectares)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(df["Production_MT"], bins=10, kde=True , color="green")
plt.title("Distribution of Production (Metric Tons)")
plt.tight_layout()
plt.show()
```

The histogram shows the distribution of production in metric tons, highlighting common production levels. Peaks indicate frequent values, while the KDE curve reveals density trends and potential outliers.

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(df["Yield_MT_per_Hectare"], bins=10, kde=True, color="red")
plt.title("Distribution of Yield (MT per Hectare)")
plt.tight_layout()
plt.show()
```

The histogram illustrates the distribution of yield per hectare, showing common yield levels. Peaks indicate frequent values, while the KDE curve highlights density trends and variations.

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(df["Value_Lakhs"], bins=10, kde=True, color="purple")
plt.title("Distribution of Value (Rs. in Lakhs)")
plt.tight_layout()
plt.show()
```

The histogram shows the distribution of value in lakhs, highlighting the most common value ranges. The KDE curve reveals density trends.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Area_Hectares", y="Production_MT", data=df, hue="Yield_MT_per_Hectare", size="Value_Lakhs", sizes=(20, 200), palette="coolwarm", edgecolor="black")
plt.xlabel("Area (Hectares)")
plt.ylabel("Production (Metric Tons)")
plt.title("Scatter Plot of Area vs. Production (with Yield and Value)")
plt.legend(title="Yield (MT/Hectare) & Value Size")
plt.show()
```

The scatter plot shows how production increases with area, with yield variations represented by color. Larger points indicate higher value, revealing differences in efficiency and economic significance across different agricultural regions.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Area_Hectares", y="District_Name", data=df.sort_values("Area_Hectares", ascending=False), palette="viridis")
plt.xlabel("Area (Hectares)")
plt.ylabel("District Name")
plt.title("Area Across Different Districts")
plt.show()
```

The bar plot shows the distribution of agricultural area across districts. Some districts have significantly larger cultivation areas, highlighting regional differences in land usage for agricultural production.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Area_Hectares", y="Value_Lakhs", data=df, hue="Production_MT", size="Yield_MT_per_Hectare", sizes=(20, 200), palette="plasma", edgecolor="black")
plt.xlabel("Area (Hectares)")
plt.ylabel("Value (Rs. in Lakhs)")
plt.title("Scatter Plot of Area vs. Value")
plt.show()
```

The scatter plot shows that higher agricultural area generally leads to higher value in lakhs. However, variations in production and yield impact overall value, with some regions achieving high value despite smaller land areas.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Yield_MT_per_Hectare", y="Value_Lakhs", data=df, hue="Area_Hectares", size="Production_MT", sizes=(20, 200), palette="magma", edgecolor="black")
plt.xlabel("Yield (MT per Hectare)")
plt.ylabel("Value (Rs. in Lakhs)")
plt.title("Scatter Plot of Yield vs. Value")
plt.show()
```

The scatter plot indicates that higher yield per hectare generally correlates with increased value in lakhs. However, variations exist, as some areas achieve high value despite lower yield, influenced by total production and land area.

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x="District_Name", y="Yield_MT_per_Hectare", data=df, palette="muted")
plt.xticks(rotation=90)
plt.xlabel("District Name")
plt.ylabel("Yield (MT per Hectare)")
plt.title("Distribution of Yield Across Districts")
plt.show()
```

The violin plot shows the distribution of yield per hectare across different districts. Some districts have a wide range of yields, while others are more consistent, indicating regional differences in agricultural productivity.

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df["Production_MT"], shade=True, color="green")
plt.title("Density Distribution of Production (Metric Tons)")
plt.xlabel("Production (Metric Tons)")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df_sorted = df.sort_values("Value_Lakhs", ascending=False)
plt.barh(df_sorted["District_Name"], df_sorted["Value_Lakhs"], color="purple")
plt.xlabel("Value (Rs. in Lakhs)")
plt.ylabel("District Name")
plt.title("Value Across Different Districts")
plt.show()
```

The bar chart highlights the distribution of value (in lakhs) across districts. Some districts contribute significantly higher value, indicating regional disparities in agricultural or economic productivity.

```python
plt.figure(figsize=(10, 8))
plt.pie(df["Production_MT"], labels=df["District_Name"], autopct="%1.1f%%", startangle=140, colors=sns.color_palette("coolwarm", len(df)))
plt.title("Share of Total Production Across Districts")
plt.axis("equal")
plt.show()
```

The pie chart shows the share of total production across districts. Some districts contribute significantly more to production, indicating higher agricultural output, while others have a relatively smaller share.

```python
plt.figure(figsize=(8, 5))
sns.ecdfplot(df["Production_MT"], color="blue")
plt.title("ECDF of Production")
plt.xlabel("Production (Metric Tons)")
plt.ylabel("Proportion")
plt.show()
```

The ECDF plot shows the cumulative distribution of production across districts. It helps visualize how production values are spread, indicating that a majority of districts have lower production, while fewer districts have very high production.

The polar plot represents Production (MT) in a circular format. Each bar's height shows production levels, with colors from the viridis palette. This helps compare values in a radial layout, making trends and differences visually distinct across districts.

```python
plt.figure(figsize=(12, 6))
sns.stripplot(y="District_Name", x="Yield_MT_per_Hectare", data=df, jitter=True, palette="coolwarm")
plt.title("Dot Plot of Yield Across Districts")
plt.show()
```

The dot plot shows the distribution of yield (MT per hectare) across districts. It highlights variations in yield, with some districts having significantly higher or lower values, indicating regional differences in productivity.

```python
plt.figure(figsize=(12, 6))
plt.stem(df["District_Name"], df["Area_Hectares"], linefmt="r-", markerfmt="ro", basefmt="k-")
plt.xticks(rotation=90)
plt.xlabel("District Name")
plt.ylabel("Area (Hectares)")
plt.title("Lollipop Chart of Area")
plt.show()
```

The lollipop chart visualizes the area (in hectares) across different districts. It highlights the variations in land area, with some districts having significantly larger agricultural land compared to others, showcasing regional differences.
