```python
import pandas as pd
df=pd.read_csv("class_marks.csv")
a=df.sort_values("Total",ascending=True)
a
```

```python
df.fillna(0,inplace=True)
df['Q1'] = df["Q1aM4"] + df["Q1bM6"]
df['Q2'] = df['Q2aM6'] + df['Q2bM4']
df['Q3'] = df['Q3aM5'] + df['Q3bM5']
df['Q4'] = df['Q4aM3'] + df['Q4bM7']
df['Q5']=df['Q5M10']
df['Q6'] = df['Q6aM4'] + df['Q6bM6']
df.drop(["Q1aM4","Q1bM6",'Q2aM6','Q2bM4','Q3aM5','Q3bM5','Q4aM3','Q4bM7','Q5M10','Q6aM4','Q6bM6'],axis=1,inplace=True)
```

```python
a=df.loc[(df['Total'] >= 15) & (df['Total'] <= 20)]
a=a.reset_index()
a
```

```python
b=a.groupby('Q1')['Total']
b.hist(color = 'blue',figsize=[6,6],grid=False,bins=5)
plt.title("Histogram of students b/w 15-20 Marks")
plt.xlabel("Total Marks obtained")
plt.ylabel("Marks Q1")
plt.show()
```

The histogram shows the distribution of students' total marks for Q1, highlighting how many scored between 15-20 marks in different bins.

```python
a6=a.groupby('Q6')['Total']
a6.hist(color='orange',figsize=[6,6]
,grid=False,bins=8)
plt.title("Histogram of students b/w 15-20 Total Marks")
plt.xlabel("Total Marks")
plt.ylabel("Marks Q6")
plt.show()
```

The histogram shows how students' total marks for Q6 are distributed, highlighting how many scored between 15-20 marks across eight bins.

```python
df.boxplot(by='Q3', column =['Total'], grid = False,color='blue',figsize=[6,6])
plt.title("Marks")
plt.ylabel("Total")
plt.show()
```
